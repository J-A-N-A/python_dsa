# **RAT IN A MAZE PROBLEM**

maze.jpg
So, here's what you probably did - you took a path, kept going, you hit an obstacle, you came back all the way long to the start point, you took a different path and kept going. And, the cycle repeats until you find the right path that's gonna lead you to the end point, isn't it ?

Well, this simple technique of problem solving is what we call "Backtracking Technique" in programming.

The idea behind backtracking is pretty simple  - We try to build a solution step by step using recursion; if during the process we face any obstacles or realize that it is not going to be a valid solution, then we stop computing that solution and we return back to the step before, i.e we simply backtrack.



Backtracking is a pretty simple, straightforward and an elegant solution. Let's just begin with a simple challenge to understand things behind the technique.

Broadly, here's the logic we are gonna be using - Every time the rat reaches a particular cell in the matrix (maze), we are gonna be checking for 3 conditions:



Is it the end point of the matrix i.e (3, 3)
Does it meet the boundary conditions
Is the cell value 1

If the particular cell meets the above criteria, then we are gonna mark it as a path and store it in the solution matrix of same size and further continue traversing recursively downwards. Else, we will backtrack, and traverse horizontally.
Notes to remember :



Here's an overview of the logic to solve the Rat in a maze challenge.



If destination is reached <br>
  print the solution matrix<br>
Else<br>
  a. Mark current cell in the solution matrix<br>
  b. Move down and recursively <br>
check if this leads to a solution<br>
  c. If there is no solution, trace back and move horizontally and recursively checking if this leads to a solution
  d. If none of the above solutions work, unmark the cell and return False

Notes to remember :



Our objective is to place N queens on an NxN chessboard such that they don't clash with each other. Now, what makes this simple problem real hard is the fact that a queen can typically move in all possible directions - horizontally , vertically  and even diagonally .



With the use of backtracking, the best possible approach we have right in front of us is:

Start at the leftmost column i.e 1st row, 1st column.
The first queen can obviously be placed without any clashes. So, mark this [row, column] as a part of the soultion.
Now, bring in the second queen. Place it in the first column, second row. Then, check if the queen is safe in there.
If the 2nd queen isn't safe in it's current position, then move it horizontally as long as it reaches a safe position.
If you are able to find out a safe position for the 2nd queen in 2nd row, then, bring in the 3rd queen into the picture and keep going.
If you are not able to identify a safe position for the 2nd queen, then backtrack i.e go back to the 1st queen and change it's position. Subsequently, check if 2nd queen is atleast now able to find a safer position .



  
And, this goes on until all N queens reach their positions
