##**DIVIDE AND CONQUER**

ÔªøÔªøÔªøÔªøÔªøDivide and conquer is based on multi-branched recursion. The basic idea behind this design paradigm is:



Divide: If the problem is small, then solve it directly. Otherwise, divide the problem into smaller subsets of the same problem. Divide until the small subset becomes simple enough to solve directly.
Conquer: Conquer the smaller subsets by solving them recursively. 
Combine: Take the solutions of the sub-problems and merge them to get a solution to the original problem.

```python
#EXAMPLE CODE TO FIND POWER OF A NUMBER USING DIVIDE AND CONQUER APPROACH 
#time complexity o(b)
def power(a,b):
  if(b==0):
  	return 1
  elif(int(b % 2) == 0):                             #CHECK IF IT A EVEN POWER 
    return (power(a, int(b/2)) * power(a, int(b / 2)))  #IF EVEN POWER SPLIT IT BY 2
  else:                                                 
    return (a * power(a ,int(b/2)) * power(a, int(b/2))) #ELSE IF IT IS ODD DIVIDE IT INTO 2 AND MULTIPLY WITH a itself
#driver code
a=2
b=4
print(power(a,b))
```
