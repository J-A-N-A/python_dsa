# **SORTING**
BUBBLE **SORT**
Notes to remember:



A bubble sort algorithm iterates through the list that it is given, and compares each pair of adjacent elements in the list. If the elements are in the incorrect order, it swaps them, and then moves on to the next pair of adjacent elements.

for i = 0 to n - 1 // to keep track of number of cycles<BR>
for j = 0 to n - i - 1 // to compare the elements within an iteration<BR>
// swap if one element is greater than its adjacent element<BR>
if arr[j] > arr[j + 1] then<BR>
swap(arr[j], arr[j + 1])<BR>
end if<BR>
end for<BR>
end for

**Time complexity**:  O(n^2) average, worst. O(n) best if input is already sorted.<BR>
Quick Visualization : 
71784_1580551129.png

```python
#implementation of bubble sort for an unsorted array
def bubble(arr):
  n=len(arr)
  for i in range(0,n-1):
    for j in range(0,n-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
arr=[1,23,4,56,87]
bubble(arr)
for i in range(len(arr)):
    print ("% d" % arr[i],end=" ")
```

## **INSERTION SORT**
Notes to remember :



Insertion sort algorithm works this way: it maintains two subsets  a sorted subset, and an unsorted subset. It inserts an element from an unsorted subset to the sorted subset until all the elements are sorted.


Pseudocode:

For each value index 1 to n - 1 <br>
Compare with all elements to the left of the current value to determine new insertion point <br>
Hold current value in temp variable <br>
Shift elements from new insertion point right<br>
Insert value in temp variable<br>
Break<br>

**Time complexity**: O(n^2) average, worst. O(n) best if input is already sorted
74946_1580621974.png

```python
#implementation of insertion sort

def isort(arr):
  for step in range(1,len(arr)):
    key=arr[step]
    j=step-1
    while j >=0 and key<arr[j]:  
      arr[j+1]=arr[j]
      j=j-1
    arr[j+1]=key
arr=[1,23,5,6,7,8]
isort(arr)
print(arr)
```

## **SELECTION SORT**
Notes to remember :



A Selection sort algorithm sorts a list of elements by iterating and finding the smallest/largest one and then putting it aside into a sorted list. This happens until there are no elements left in the unsorted list.<BR>

**PSEUDOCODE<BR>**

For each element<BR>
Check every element to the right to find the min<BR>
If min < current element, swap<BR>

**Time Complexity**: O(n^2) average, worst, best

```python
#implementation of selection sort
def ssort(arr):
    for i in range(0,len(arr)):
      minidx=i
      for j in range(i+1,len(arr)):
    	  if arr[minidx]>arr[j]:
           minidx=j
      arr[i],arr[minidx]=arr[minidx],arr[i]
arr=[1,2,3,4,5,6,3]
ssort(arr)
print(arr)
      
```
